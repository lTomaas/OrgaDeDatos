```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
reviews = pd.read_csv('C:/Users/tomas/OneDrive/Escritorio/ORGA DE DATOS/review.csv')
```

```python
datos = reviews[["useful","funny","cool"]]
```

```python
datos_corr = datos.corr()
datos_corr
```

```python
plt.figure(dpi = 200)
sns.heatmap(datos_corr, xticklabels = datos_corr.columns, yticklabels = datos_corr.columns, cmap = 'Greys',vmin = 0.75,vmax = 1)
plt.title("Correlaci√≥n entre votos useful, funny y cool")
```
