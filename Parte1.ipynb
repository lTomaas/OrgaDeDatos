```python
import pandas as pd
import pyarrow.parquet as pq 
import matplotlib.pyplot as plt
import seaborn as sns
train = pq.read_table(source='C:/Users/tomas/OneDrive/Escritorio/ORGA DE DATOS/train').to_pandas().dropna()
```

```python
plt.figure(dpi = 150)
sns.violinplot(data = train, x = "popular", y = "timedelta", orient = "v", palette="Greens")
plt.title("Distribucion de popularidad en funcion de la antiguedad")
plt.xlabel("Es popular")
plt.xticks([0,1], ["No","Si"])
plt.ylabel("Antiguedad en dias")
```

```python
plt.figure(dpi = 800)
plt.xticks(rotation=0, size = 15.5)
sns.barplot(data = train, y = "num_videos", x = "popular",ci=None, palette="Blues_d")
plt.title("Distribucion de popularidad en funcion de la mediana de videos")
plt.xlabel("Es popular")
plt.xticks([0,1], ["No","Si"])
plt.ylabel("Mediana de videos")
```

```python
matrix = pd.pivot_table(train, values = "num_imgs", columns = "data_channel_is_world", index = "popular", aggfunc= "mean")
plt.figure(dpi = 300)
sns.heatmap(matrix, cmap = "Blues", vmin = 0, vmax = 7, square = True,linewidths = .5, annot = True)
plt.title("Mediana de imagenes en funcion de\nsi es popular y si el canal trata del mundo")
plt.xlabel("El canal trata del mundo")
plt.xticks([0.5,1.5], ["No","Si"])
plt.ylabel("Es popular")
plt.yticks([0.5,1.5], ["No","Si"])
```

```python
plt.figure(dpi = 150)
plt.title("Precio de venta segun la condicion de las calles")
sns.boxplot(data = train, x = "popular", y = "num_hrefs", showfliers = False, palette="Purples")
plt.title("Distribucion de popularidad en funcion de la cantidad de links")
plt.ylabel("Cantidad de links")
plt.xlabel("Es popular")
plt.xticks([0,1], ["No","Si"])
```

```python
plt.figure(dpi = 800)
plt.xticks(rotation=0, size = 15.5)
sns.barplot(data = train, y = "num_imgs", x = "popular",ci=None, palette="Greys")
plt.title("Distribucion de popularidad en funcion de la mediana de imagenes")
plt.xlabel("Es popular")
plt.xticks([0,1], ["No","Si"])
plt.ylabel("Mediana de imagenes")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(dpi = 150)
sns.violinplot(data = train, x = "popular", y = "title_subjectivity", orient = "v", palette="Reds")
plt.title("Distribucion de popularidad en funcion de la subjetividad del titulo")
plt.xlabel("Es popular")
plt.xticks([0,1], ["No","Si"])
plt.ylabel("Subjetividad del titulo")
```
