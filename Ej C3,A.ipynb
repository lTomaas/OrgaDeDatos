```python
import zstandard
def compress_size(bytefile):
    return len(zstandard.compress((bytefile.encode())))


def ncd(st1,st2):
    len1 = compress_size(st1)
    len2 = compress_size(st2)
    concat_len =  compress_size(st1+st2)
    return (concat_len - min(len1,len2))/max(len1,len2)
```
