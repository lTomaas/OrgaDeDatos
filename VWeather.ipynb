```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df =  pd.read_csv("C:/Users/tomas/OneDrive/Escritorio/ORGA DE DATOS/weatherAUS.csv")
df
```

```python
plt.figure(dpi = 800)
plt.xticks(rotation=90, size = 5.5)
sns.barplot(data = df, y = "Rainfall", x = "WindGustSpeed",ci=None, palette="Blues_d")
plt.title("Distribucion de las precipitaciones en funcion de la velocidad del viento")
plt.xlabel("Velocidad del viento")
plt.ylabel("Precipitacion")
```

```python
plt.figure(dpi = 300)
plt.xticks(rotation = 90, size = 15) 
sns.boxplot(data = df, x = "WindGustDir", y = "Rainfall", showfliers = False,color= "Grey")
plt.title("Precipitacion en funcion de la direccion del viento")
plt.ylabel("Precipitacion")
plt.xlabel("Direccion del viento")
```
