```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
archivo = pd.read_csv('C:/Users/tomas/OneDrive/Escritorio/ORGA DE DATOS/train.csv')
archivo
```

```python
plt.figure(dpi = 150)
plt.hist(archivo.SalePrice)
plt.title("Distribucion de indice del precio de venta")
plt.xlabel("Precio de venta")
plt.ylabel("Cantidad")
```

```python
plt.figure(dpi = 150)
plt.title("Precio de venta segun la condicion de las calles")
sns.boxplot(data = archivo, x = "Street", y = "SalePrice", showfliers = False)
plt.ylabel("Precio de venta")
plt.xlabel("Condicion de calle")
```
