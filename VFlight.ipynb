```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df =  pd.read_csv("C:/Users/tomas/OneDrive/Escritorio/ORGA DE DATOS/flight_delays_train.csv")
df
```

```python
matrix = pd.pivot_table(df, values = "DepTime", columns = "UniqueCarrier", index = "dep_delayed_15min", aggfunc= "mean")
plt.figure(dpi = 300)
sns.heatmap(matrix, cmap = "Reds", vmin = 1250, vmax = df.DepTime.mean() , square = True)
plt.title("Demora en tiempo de partida en funcion\n de la compaÃ±ia y si hubo retraso en la salida")
plt.xlabel("Codigo de compaÃ±ia")
plt.ylabel("Retraso en la salida")
```

```python
plt.figure(dpi = 150)
sns.violinplot(data = df, y = "DepTime", x = "dep_delayed_15min", orient = "v", palette="Reds")
plt.title("Distribucion de la hora de salida en funcion del retraso")
plt.xlabel("Demora en la salida")
plt.ylabel("Hora de salida")
```
